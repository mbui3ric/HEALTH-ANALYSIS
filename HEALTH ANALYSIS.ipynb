```python
#BINS FOR VARIOUS AGE CATEGORIES:
```

```python
young = df["age"].quantile(0.25)#find the 25th percentile(q1) of the age column
mid_age = df["age"].median()#find the  median value in the age column
senior = df["age"].quantile(0.75)#find the 75th percentile(q3) of the age column
elder = df["age"].max()#find the maximum(highest)value in the age column
```

```python
#DEFINE THE LABELS FOR EACH OF THE BINS
bin_labels =["Young_Adult","Middle_Age","Senior_Citizen","Elderly"]
bins = [0,young,mid_age,senior,elder]
```

```python
#CREATE A NEW COLUMN "Age_Category" BASED ON THE BINS AND LABELS
df["Age_Category"] = pd.cut(df["age"],labels = bin_labels,bins = bins)
df["Age_Category"]
```

```python
#VISUALIZATION OF THE ABOVE DATA
```

```python
columns = ["chest pains","Age_Category","Sex_Category","ExAng_Category"]
def plot_cat_count():
    for c in columns:
        plt.figure(figsize=(10,6))
        fig = sns.countplot(x = c,data = df,palette = "dark")
        plt.title=(c)
        plt.ylabel=("Count")
        plt.xlabel=(c)
        plt.tight_layout()
        #ADD COUNT VALUES ON TOP OF EACH COLUMN
        for bars in fig.containers:
            fig.bar_label(bars)
        plt.show()
plot_cat_count()
```
